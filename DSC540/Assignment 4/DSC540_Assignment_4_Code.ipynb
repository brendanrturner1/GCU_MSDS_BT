```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
from sklearn import svm
from sklearn.metrics import mean_squared_error
#Import packages
```

```python
rd.seed(17)
X=np.linspace(-3,3,50) #Generate 50 random points in the space [-3,3]
err=np.random.normal(0,1,50) #Create random Gaussian noise 
X=X+err #Add the noise to the data
y=np.sinc(X) #Use sinc function for testing algorithm
```

```python
rbf_svr=svm.SVR(kernel='rbf',C=100,gamma=0.5)
#Define the SVM Regressor
```

```python
X=X.reshape(-1,1)
mod=rbf_svr.fit(X,y)
#Fit the regressor to the data
```

```python
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape(-1,1)
plt.scatter(X,y)
plt.plot(X_grid,mod.predict(X_grid))
#Plot the fit of the model
```

```python
rbf_svr2=svm.SVR(kernel='rbf',C=100,gamma=0.1,epsilon=0.1)
#Changing to a smaller gamma value, to show impact 
mod2=rbf_svr2.fit(X,y)
plt.scatter(X,y)
plt.plot(X_grid,mod2.predict(X_grid))
```

```python
rbf_svr3=svm.SVR(kernel='rbf',C=.1,gamma=0.5)
#Changing to a smaller C value, to show impact 
mod3=rbf_svr3.fit(X,y)
plt.scatter(X,y)
plt.plot(X_grid,mod3.predict(X_grid))
```

```python
rbf_svr4=svm.SVR(kernel='rbf',C=10000000,gamma=0.1)
#Changing to a very large C value and small gamma, to show impact 
mod4=rbf_svr4.fit(X,y)
plt.scatter(X,y)
plt.plot(X_grid,mod4.predict(X_grid))
```

```python
mean_squared_error(X_grid,mod4.predict(X_grid))
```
