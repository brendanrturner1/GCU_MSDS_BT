```python
x1=np.linspace(-10,10,500) 
x2=np.linspace(-10,10,500)
y = (sin(x1) * sin(x2)) / (x1 * x2)

# Training data are sampled uniformly from the input range [–10, 10] × [–10, 10]. With two
# symmetric triangular membership functions assigned to each input variable, construct a
# Takagi-Sugeno fuzzy model (linear static mappings) for the sinc function. Provide defining
# equations for determination of the premise and consequent parameters of the model.
```

```python
from simpful import *
# A simple fuzzy inference system for the tipping problem
# Create a fuzzy system object
FS = FuzzySystem()

# Define fuzzy sets and linguistic variables
A_1 = FuzzySet(points=[[-3., 0.], [0., 1.], [3., 0.]], term="big1")
A_2 = FuzzySet(points=[[-10., 1.], [-2., 0.], [2., 0.], [10., 1.]], term="small1")
FS.add_linguistic_variable("x1", LinguisticVariable([A_1, A_2], concept="x1 Q"))

B_1 = FuzzySet(points=[[-3., 0.], [0., 1.], [3., 0.]], term="big2")
B_2 = FuzzySet(points=[[-10., 1.], [-2., 0.], [2., 0.], [10., 1.]], term="small2")
FS.add_linguistic_variable("x2", LinguisticVariable([B_1, B_2], concept="x2 Q"))

# Define output crisp values
FS.set_crisp_output_value("small", 5)
FS.set_crisp_output_value("medium", 10)

# Define function for large
FS.set_output_function("large", "x1+x2+5")

# Define fuzzy rules
R1 = "IF (x1 IS big1) AND (x2 IS big2) THEN (y IS large)"
R2 = "IF (x1 IS small1) AND (x2 IS big2) THEN (y IS medium)"
R3 = "IF (x1 IS small1) AND (x2 IS small2) THEN (y IS small)"
FS.add_rules([R1, R2, R3])

# Set antecedents values
FS.set_variable("x1", 5)
FS.set_variable("x2", 5)

# Perform Sugeno inference and print output
print(FS.Sugeno_inference(["y"]))
```
