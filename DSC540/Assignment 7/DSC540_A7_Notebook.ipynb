```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import csv

data = pd.read_csv(r"C:\Users\giant\Downloads\Mall_Customers.csv")
```

```python
data.head()
```

```python
data=data.rename(columns={'CustomerID':'ID','Annual Income (k$)':'Income','Spending Score (1-100)': 'Score','Genre':'Sex'})
```

```python
data['Sex'] = data.Sex.map({'Female':0, 'Male':1})
```

```python
plt.hist(data.Score, 12)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data.Income, 12)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
X=data.iloc[:,3:].values
```

```python
c=15
wcss = []
for i in range(1,c):
    m=KMeans(i,random_state=7)
    m.fit(X)
    wcss.append(m.inertia_)
plt.plot(range(1,c),wcss,'o-',color='red')
plt.xlabel('Numbers of Clusters')
plt.ylabel('WCSS Score')
plt.title("Elbow Method")
plt.show()
```

```python
mk=KMeans(5,random_state=77)
mk_pred=mk.fit_predict(X)
plt.figure(figsize=(8,7))
colors=['r','b','g','y','pink']
for i in range(5):
    plt.scatter(X[mk_pred==i,0],X[mk_pred==i,1],s=50,c=colors[i])
plt.scatter(mk.cluster_centers_[:,0],mk.cluster_centers_[:,1],color='Black',marker='+',s=300)
plt.title("K-Means Clustering for Mall Customer Data, K=5 Clusters")
plt.show()
```

```python
# For comparison, here is K=3 clusters

mk=KMeans(3,random_state=77)
mk_pred=mk.fit_predict(X)
plt.figure(figsize=(8,7))
colors=['r','b','g','y','pink']
for i in range(3):
    plt.scatter(X[mk_pred==i,0],X[mk_pred==i,1],s=50,c=colors[i])
plt.scatter(mk.cluster_centers_[:,0],mk.cluster_centers_[:,1],color='Black',marker='+',s=300)
plt.title("K-Means Clustering for Mall Customer Data, K=3 Clusters")
plt.show()
```
